```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data_LinearRegression.csv")
X,Y = df.x, df.y
```

```python
df
```

```python
n = len(X)
epochs = 100
learning_rate = 0.0001

k,b = 0, 1
y_pred = k*X + b
cost = sum((Y - y_pred)**2)/n

for i in range(epochs):
  if(i == 0):
    plt.figure(figsize=(15, 6))
    plt.scatter(X,Y)
    plt.plot(X, k*X + b, color = "r")
    plt.title(f"Line on epoch number {i}")
    plt.show()

  k_gradient = (-2/n)*sum(X*(Y-y_pred))
  b_gradient = (-2/n)*sum(Y-y_pred)

  k -= k_gradient*learning_rate
  b -= b_gradient*learning_rate
  
  y_pred = k*X + b
  old_cost = cost
  cost = sum((Y - y_pred)**2)/n

  if(old_cost - cost < 0.0001):
    plt.figure(figsize=(15, 6))
    plt.scatter(X,Y)
    plt.plot(X, k*X + b, color = "r")
    plt.show()
    print(f"epoch number {i} was good enough")
    break
```
